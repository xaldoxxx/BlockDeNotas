```python
import mido
from mido import MidiFile, MidiTrack, Message
from decimal import Decimal, getcontext

class GoldenSpiralMusicGenerator:
    def __init__(self, length, phi):
        self.length = length
        self.phi = phi
        self.scale = ['C', 'D', 'E', 'F', 'G', 'A', 'B']

    def generate_sequence(self):
        sequence = [0, 1]
        for i in range(2, self.length):
            sequence.append(sequence[i - 1] + sequence[i - 2])
        return sequence

    def map_to_notes(self, sequence):
        notes = []
        for value in sequence:
            note_index = int(Decimal(value).quantize(Decimal('1')).to_integral()) % len(self.scale)
            note = self.scale[note_index]
            notes.append(note)
        return notes

    def generate_midi(self, filename, bpm=120):
        sequence = self.generate_sequence()
        notes = self.map_to_notes(sequence)

        midi = MidiFile()
        track = MidiTrack()
        midi.tracks.append(track)

        ticks_per_beat = mido.second2tick(1, ticks_per_beat=480, tempo=bpm)

        for note in notes:
            track.append(Message('note_on', note=note, velocity=64, time=0))
            track.append(Message('note_off', note=note, velocity=64, time=ticks_per_beat))

        midi.save(filename)


# Configuración para mostrar el resultado en formato decimal normal
getcontext().prec = 50

# Ejemplo de uso
length = 20  # Longitud de la secuencia musical
phi = Decimal('1.6180339887')  # Número áureo (phi)

music_generator = GoldenSpiralMusicGenerator(length, phi)
sequence = music_generator.generate_sequence()
notes = music_generator.map_to_notes(sequence)

print("Secuencia generada:", sequence)
print("Notas mapeadas:", notes)

# Generar archivo MIDI
filename = 'golden_spiral_music.mid'
music_generator.generate_midi(filename)
print("Archivo MIDI generado:", filename)

```

```python
from music21 import chord, stream

class ChordProgressionGenerator:
    def __init__(self, root_note):
        self.root_note = root_note
        self.chord_progression = []

    def generate_chord_progression(self, spiral_sequence, ted_greene_system):
        for i in range(len(spiral_sequence)):
            chord_name = ted_greene_system[i % len(ted_greene_system)]
            chord = self.create_chord(chord_name)
            self.chord_progression.append(chord)

        return self.chord_progression

    def create_chord(self, chord_name):
        full_chord_name = self.root_note + chord_name
        chord_obj = chord.Chord(full_chord_name)

        return chord_obj

    def create_midi_stream(self):
        midi_stream = stream.Stream()
        for chord in self.chord_progression:
            midi_stream.append(chord)

        return midi_stream

# Definir la secuencia de la espiral de oro y el sistema V de Ted Greene
spiral_sequence = [0, 1, 1, 2, 3, 5, 8, 13]
ted_greene_system = ['7', 'm7', 'm7b5', '7sus4', 'maj7', 'm7', 'm7b5', '7']

# Crear generador de progresión de acordes
progression_generator = ChordProgressionGenerator('G')
chord_progression = progression_generator.generate_chord_progression(spiral_sequence, ted_greene_system)

# Crear stream MIDI
midi_stream = progression_generator.create_midi_stream()

# Guardar archivo MIDI
filename = 'golden_spiral_blues.mid'
midi_stream.write('midi', fp=filename)
print("Archivo MIDI generado:", filename)
```

```python
class PrimeGenerator:
    def __init__(self):
        self.primes = []

    def generate_primes(self, start, end):
        self.primes = []
        for num in range(start, end + 1):
            if self.is_prime(num):
                self.primes.append(num)

    def is_prime(self, num):
        if num < 2:
            return False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True

    def find_twin_primes(self):
        twin_primes = []
        for i in range(len(self.primes) - 1):
            if self.primes[i + 1] - self.primes[i] == 2:
                twin_primes.append((self.primes[i], self.primes[i + 1]))
        return twin_primes


# Uso del generador de números primos
prime_generator = PrimeGenerator()
prime_generator.generate_primes(1, 1000000)
twin_primes = prime_generator.find_twin_primes()

print("Números primos gemelos encontrados:")
for twin_prime in twin_primes:
    print(twin_prime)
```

```python
class MersennePrimeGenerator:
    def __init__(self):
        self.mersenne_primes = []

    def generate_mersenne_primes(self, start, end):
        self.mersenne_primes = []
        for num in range(start, end + 1):
            if self.is_prime(num):
                mersenne_num = 2**num - 1
                if self.is_prime(mersenne_num):
                    self.mersenne_primes.append(mersenne_num)

    def is_prime(self, num):
        if num < 2:
            return False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True


# Uso del generador de números primos de Mersenne
mersenne_prime_generator = MersennePrimeGenerator()
mersenne_prime_generator.generate_mersenne_primes(1, 10)

print("Números primos de Mersenne encontrados:")
print(mersenne_prime_generator.mersenne_primes)
```

```python
class AmicableNumberGenerator:
    def __init__(self):
        self.amicable_numbers = []

    def generate_amicable_numbers(self, start, end):
        self.amicable_numbers = []
        for num in range(start, end + 1):
            if self.is_amicable(num):
                self.amicable_numbers.append(num)

    def is_amicable(self, num):
        divisor_sum1 = self.calculate_divisor_sum(num)
        if divisor_sum1 == num:
            return False
        divisor_sum2 = self.calculate_divisor_sum(divisor_sum1)
        return divisor_sum2 == num

    def calculate_divisor_sum(self, num):
        divisor_sum = 0
        for i in range(1, num):
            if num % i == 0:
                divisor_sum += i
        return divisor_sum


# Uso del generador de números amigos
amicable_number_generator = AmicableNumberGenerator()
amicable_number_generator.generate_amicable_numbers(1, 10000)

print("Números amigos encontrados:")
print(amicable_number_generator.amicable_numbers)
```

```python
class PerfectNumberGenerator:
    """
    Clase que genera números perfectos en un rango dado.
    """

    def __init__(self):
        self.perfect_numbers = []

    def generate_perfect_numbers(self, start, end):
        """
        Genera números perfectos en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.perfect_numbers = []
        for num in range(start, end + 1):
            if self.is_perfect(num):
                self.perfect_numbers.append(num)

    def is_perfect(self, num):
        """
        Verifica si un número dado es perfecto.

        :param num: Número a verificar.
        :return: True si el número es perfecto, False en caso contrario.
        """
        divisors = self.get_proper_divisors(num)
        divisor_sum = sum(divisors)
        return divisor_sum == num

    def get_proper_divisors(self, num):
        """
        Obtiene los divisores propios de un número dado.

        :param num: Número del cual se obtendrán los divisores propios.
        :return: Lista de divisores propios del número.
        """
        divisors = []
        for i in range(1, num):
            if num % i == 0:
                divisors.append(i)
        return divisors


# Uso del generador de números perfectos
perfect_number_generator = PerfectNumberGenerator()
perfect_number_generator.generate_perfect_numbers(1, 10000)

print("Números perfectos encontrados:")
print(perfect_number_generator.perfect_numbers)
```

```python
class SociableNumberGenerator:
    """
    Clase que genera números sociables en un rango dado.
    """

    def __init__(self):
        self.sociable_numbers = []

    def generate_sociable_numbers(self, start, end):
        """
        Genera números sociables en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.sociable_numbers = []
        for num in range(start, end + 1):
            if self.is_sociable(num):
                self.sociable_numbers.append(num)

    def is_sociable(self, num):
        """
        Verifica si un número dado es sociable.

        :param num: Número a verificar.
        :return: True si el número es sociable, False en caso contrario.
        """
        chain = [num]
        while True:
            next_num = self.sum_proper_divisors(chain[-1])
            if next_num == num:
                return True
            elif next_num in chain:
                return False
            else:
                chain.append(next_num)

    def sum_proper_divisors(self, num):
        """
        Calcula la suma de los divisores propios de un número dado.

        :param num: Número del cual se calculará la suma de los divisores propios.
        :return: Suma de los divisores propios.
        """
        divisors = self.get_proper_divisors(num)
        return sum(divisors)

    def get_proper_divisors(self, num):
        """
        Obtiene los divisores propios de un número dado.

        :param num: Número del cual se obtendrán los divisores propios.
        :return: Lista de divisores propios del número.
        """
        divisors = []
        for i in range(1, num):
            if num % i == 0:
                divisors.append(i)
        return divisors


# Uso del generador de números sociables
sociable_number_generator = SociableNumberGenerator()
sociable_number_generator.generate_sociable_numbers(1, 100)

print("Números sociables encontrados:")
print(sociable_number_generator.sociable_numbers)
```

```python
class OmirpNumberGenerator:
    """
    Clase que genera números omirp en un rango dado.
    """

    def __init__(self):
        self.omirp_numbers = []

    def generate_omirp_numbers(self, start, end):
        """
        Genera números omirp en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.omirp_numbers = []
        for num in range(start, end + 1):
            if self.is_omirp(num):
                self.omirp_numbers.append(num)

    def is_omirp(self, num):
        """
        Verifica si un número dado es omirp.

        :param num: Número a verificar.
        :return: True si el número es omirp, False en caso contrario.
        """
        reversed_num = int(str(num)[::-1])
        if self.is_prime(num) and self.is_prime(reversed_num) and num != reversed_num:
            return True
        return False

    def is_prime(self, num):
        """
        Verifica si un número dado es primo.

        :param num: Número a verificar.
        :return: True si el número es primo, False en caso contrario.
        """
        if num < 2:
            return False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True


# Uso del generador de números omirp
omirp_number_generator = OmirpNumberGenerator()
omirp_number_generator.generate_omirp_numbers(1, 10000)

print("Números omirp encontrados:")
print(omirp_number_generator.omirp_numbers)
```

```python
class VampireNumberGenerator:
    """
    Clase que genera números Vampiro en un rango dado.
    """

    def __init__(self):
        self.vampire_numbers = []

    def generate_vampire_numbers(self, start, end):
        """
        Genera números Vampiro en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.vampire_numbers = []
        for num in range(start, end + 1):
            if self.is_vampire(num):
                self.vampire_numbers.append(num)

    def is_vampire(self, num):
        """
        Verifica si un número dado es Vampiro.

        :param num: Número a verificar.
        :return: True si el número es Vampiro, False en caso contrario.
        """
        digits = str(num)
        num_digits = len(digits)

        # Comprobamos que el número tiene una cantidad par de dígitos
        if num_digits % 2 != 0:
            return False

        # Generamos todas las posibles combinaciones de dígitos del número
        combinations = self.generate_digit_combinations(digits)

        # Verificamos si alguna combinación cumple la propiedad de ser un Vampiro
        for combo in combinations:
            fangs = self.get_fangs(combo)
            if fangs and fangs[0] * fangs[1] == num:
                return True

        return False

    def generate_digit_combinations(self, digits):
        """
        Genera todas las posibles combinaciones de dígitos del número.

        :param digits: Dígitos del número.
        :return: Lista de todas las combinaciones posibles.
        """
        from itertools import permutations
        return permutations(digits)

    def get_fangs(self, combo):
        """
        Obtiene los "colmillos" o "fangs" del número Vampiro.

        :param combo: Combinación de dígitos.
        :return: Tupla con los colmillos si es una combinación válida, None en caso contrario.
        """
        x = int(''.join(combo[:len(combo)//2]))
        y = int(''.join(combo[len(combo)//2:]))
        if (x % 10 == 0 and y % 10 == 0) or (x % 10 != 0 and y % 10 != 0):
            return None
        return x, y


# Uso del generador de números Vampiro
vampire_number_generator = VampireNumberGenerator()
vampire_number_generator.generate_vampire_numbers(1000, 10000)

print("Números Vampiro encontrados:")
print(vampire_number_generator.vampire_numbers)
```

```python
class AmbitiousNumberGenerator:
    """
    Clase que genera números ambiciosos en un rango dado.
    """

    def __init__(self):
        self.ambitious_numbers = []

    def generate_ambitious_numbers(self, start, end):
        """
        Genera números ambiciosos en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.ambitious_numbers = []
        for num in range(start, end + 1):
            if self.is_ambitious(num):
                self.ambitious_numbers.append(num)

    def is_ambitious(self, num):
        """
        Verifica si un número dado es ambicioso.

        :param num: Número a verificar.
        :return: True si el número es ambicioso, False en caso contrario.
        """
        divisors = self.get_proper_divisors(num)
        divisor_sum = sum(divisors)
        return num < divisor_sum

    def get_proper_divisors(self, num):
        """
        Obtiene los divisores propios de un número dado.

        :param num: Número del cual se obtendrán los divisores.
        :return: Lista de divisores propios del número.
        """
        divisors = []
        for i in range(1, num):
            if num % i == 0:
                divisors.append(i)
        return divisors


# Uso del generador de números ambiciosos
ambitious_number_generator = AmbitiousNumberGenerator()
ambitious_number_generator.generate_ambitious_numbers(1, 10000)

print("Números Ambiciosos encontrados:")
print(ambitious_number_generator.ambitious_numbers)
```

```python
class HappyNumberGenerator:
    """
    Clase que genera números felices en un rango dado.
    """

    def __init__(self):
        self.happy_numbers = []

    def generate_happy_numbers(self, start, end):
        """
        Genera números felices en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.happy_numbers = []
        for num in range(start, end + 1):
            if self.is_happy(num):
                self.happy_numbers.append(num)

    def is_happy(self, num):
        """
        Verifica si un número dado es feliz.

        :param num: Número a verificar.
        :return: True si el número es feliz, False en caso contrario.
        """
        visited = set()
        while num != 1 and num not in visited:
            visited.add(num)
            num = self.calculate_next_number(num)
        return num == 1

    def calculate_next_number(self, num):
        """
        Calcula el siguiente número en el proceso de verificación de felicidad.

        :param num: Número actual.
        :return: Siguiente número en el proceso.
        """
        next_num = 0
        while num > 0:
            digit = num % 10
            next_num += digit * digit
            num //= 10
        return next_num


# Uso del generador de números felices
happy_number_generator = HappyNumberGenerator()
happy_number_generator.generate_happy_numbers(1, 10000)

print("Números Felices encontrados:")
print(happy_number_generator.happy_numbers)
```

```python
class NarcissisticNumberGenerator:
    """
    Clase que genera números narcisistas en un rango dado.
    """

    def __init__(self):
        self.narcissistic_numbers = []

    def generate_narcissistic_numbers(self, start, end):
        """
        Genera números narcisistas en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.narcissistic_numbers = []
        for num in range(start, end + 1):
            if self.is_narcissistic(num):
                self.narcissistic_numbers.append(num)

    def is_narcissistic(self, num):
        """
        Verifica si un número dado es narcisista.

        :param num: Número a verificar.
        :return: True si el número es narcisista, False en caso contrario.
        """
        digits = self.get_digits(num)
        num_sum = sum([digit ** len(digits) for digit in digits])
        return num == num_sum

    def get_digits(self, num):
        """
        Obtiene los dígitos de un número dado.

        :param num: Número del cual se obtendrán los dígitos.
        :return: Lista de dígitos del número.
        """
        return [int(digit) for digit in str(num)]


# Uso del generador de números narcisistas
narcissistic_number_generator = NarcissisticNumberGenerator()
narcissistic_number_generator.generate_narcissistic_numbers(1, 10000)

print("Números Narcisistas encontrados:")
print(narcissistic_number_generator.narcissistic_numbers)
```

```python
class AbundantNumberGenerator:
    """
    Clase que genera números abundantes en un rango dado.
    """

    def __init__(self):
        self.abundant_numbers = []

    def generate_abundant_numbers(self, start, end):
        """
        Genera números abundantes en el rango dado.

        :param start: Número inicial del rango.
        :param end: Número final del rango.
        """
        self.abundant_numbers = []
        for num in range(start, end + 1):
            if self.is_abundant(num):
                self.abundant_numbers.append(num)

    def is_abundant(self, num):
        """
        Verifica si un número dado es abundante.

        :param num: Número a verificar.
        :return: True si el número es abundante, False en caso contrario.
        """
        divisors_sum = sum(self.get_divisors(num))
        return divisors_sum > num

    def get_divisors(self, num):
        """
        Obtiene los divisores propios de un número dado.

        :param num: Número del cual se obtendrán los divisores.
        :return: Lista de divisores propios del número.
        """
        divisors = []
        for i in range(1, int(num/2) + 1):
            if num % i == 0:
                divisors.append(i)
        return divisors


# Uso del generador de números abundantes
abundant_number_generator = AbundantNumberGenerator()
abundant_number_generator.generate_abundant_numbers(1, 10000)

print("Números Abundantes encontrados:")
print(abundant_number_generator.abundant_numbers)
```
